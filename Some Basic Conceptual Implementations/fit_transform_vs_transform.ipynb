```python
import pandas as pd
df = pd.read_csv('/content/titanic.csv', usecols= ['Pclass', 'Age', 'Fare', 'Survived'])
df.head()
```

```python
df.isnull().sum()
```

```python
## inplace = True --> Dataframe is updated and it will return nothing, but by default it is false meaning it will return the copy of the object which we have to place in some other object
df['Age'].fillna(df.Age.median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,1:]
X
```

```python
y = df.iloc[:,0]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_train_scaled
```

```python
X_test_scaled = scaler.transform(X_test)
X_test_scaled.reshape(-1, 1)
```

```python
## fit for train data and predict for test data
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

```python
classifier.fit(X_train_scaled,y_train)
```

```python
classifier.predict(X_test_scaled)
```
